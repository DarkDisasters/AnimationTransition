```python
import pandas as pd
import numpy as np

dfCountry = pd.read_csv('./LPI_csv/LPICountry.csv')
dfData = pd.read_csv('./LPI_csv/LPIData.csv')

dataDict = {}

countryCodeList = dfCountry['Country Code'].tolist()

IncomeGroup = ['Low income', 'Lower middle income', 'Upper middle income', 'High income']
dfIncomeGroup = dfCountry['Income Group']
LowIncomeFrame = dfCountry.loc[dfIncomeGroup == 'Low income']
LowerMiddleIncomeFrame = dfCountry.loc[dfIncomeGroup == 'Lower middle income']
UpperMiddleIncomeFrame = dfCountry.loc[dfIncomeGroup == 'Upper middle income']
HighIncomeFrame = dfCountry.loc[dfIncomeGroup == 'High income']
print('lenall', len(LowIncomeFrame)+len(LowerMiddleIncomeFrame)+len(UpperMiddleIncomeFrame)+len(HighIncomeFrame))

LowIncomeCountryCode = np.array(dfCountry.loc[dfIncomeGroup == 'Low income']['Country Code']).tolist()
LowerMiddleIncomeCountryCode = np.array(dfCountry.loc[dfIncomeGroup == 'Lower middle income']['Country Code']).tolist()
UpperMiddleIncomeCountryCode = np.array(dfCountry.loc[dfIncomeGroup == 'Upper middle income']['Country Code']).tolist()
HighIncomeCountryCode = np.array(dfCountry.loc[dfIncomeGroup == 'High income']['Country Code']).tolist()

dfCountry.loc[dfCountry['Income Group'] == 'Low income', 'Income Group'] = 1
dfCountry.loc[dfCountry['Income Group'] == 'Lower middle income', 'Income Group'] = 2
dfCountry.loc[dfCountry['Income Group'] == 'Upper middle income', 'Income Group'] = 3
dfCountry.loc[dfCountry['Income Group'] == 'High income', 'Income Group'] = 4
print(dfCountry['Income Group'])


for i in range(len(countryCodeList)):
    dataDict[countryCodeList[i]] = {}

datalist = np.array(dfData).tolist()
# print('datalist',datalist)
for i in range(len(datalist)):
    dataDict[datalist[i][1]]['IncomeGroup'] = dfCountry.loc[dfCountry['Country Code']==datalist[i][1]]['Income Group'].tolist()
    dataDict[datalist[i][1]][datalist[i][3]] = [datalist[i][4]]     #07
#     dataDict[datalist[i][1]][datalist[i][3]] = [datalist[i][5]]       #10
#     dataDict[datalist[i][1]][datalist[i][3]] = [datalist[i][6]]      #12
#     dataDict[datalist[i][1]][datalist[i][3]] = [datalist[i][7]]     #14
#     dataDict[datalist[i][1]][datalist[i][3]] = [datalist[i][8]]    #16
#     dataDict[datalist[i][1]][datalist[i][3]] = [datalist[i][9]]

print('dict length',dataDict)
print(dataDict)
# print('dict2007', dataDict)
keyList = []
headData = dataDict['AFG']
keyList.append('Country Code')
for key, value in headData.items():
    keyList.append(key)
print('keylist length', len(keyList))
# print('keylist', keyList)

processDataList = []
for key, value in dataDict.items():
    dataDict[key]['Country Code'] = [key]
#     countrycode = key
    processDataList.append(dataDict[key])
#     print(dataDict[contrycode])
#     for inkey, invalue
# print('processDataList', processDataList)
# print('len processDataList', len(processDataList))
# print('len processDataList[0]', len(processDataList[0]))

finalDataFrame = pd.DataFrame(columns=keyList)
print(processDataList[0])
# result=result.append(pd.DataFrame({'idx':[i],'degree':[degree],'weight':[weight],'diameter':[diameter]}),ignore_index=True)
for i in range(len(processDataList)):
    finalDataFrame = finalDataFrame.append(pd.DataFrame(processDataList[i]))
finalDataFrame.set_index(['Country Code'], inplace=True)
print('finaldataframe', finalDataFrame)
finalDataFrame.isnull().sum(axis=0)
finalDataFrame = finalDataFrame.drop(['LP.LPI.OVRL.RK.LB','LP.LPI.OVRL.RK.UB','LP.LPI.OVRL.XQ.LB','LP.LPI.OVRL.XQ.UB'],axis=1)
# print(finalDataFrame.shape)

for column in list(finalDataFrame.columns[finalDataFrame.isnull().sum() > 0]):
    mean_val = finalDataFrame[column].mean()
    finalDataFrame[column].fillna(mean_val, inplace=True)
finalDataFrame.isnull().sum(axis=0)
# df.to_csv('d:/Users/chen_lib/Desktop/fenci_result.txt',sep='\t',index=False)
finalDataFrame.to_csv('processData/07Value.txt', sep='\t',index=False)
```
